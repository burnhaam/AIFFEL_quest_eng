```python
# 전체 메뉴와 가격 
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]
```

```python
class Kiosk: 
    def __init__(self):
        self.menu = menu
        self.price = price

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        print("\n **** 주문을 시작합니다. ****")  # 메뉴 선택 안내

        # 주문 완료 후 주문 리스트 만들기 (축적을 위해 루프 밖으로 이동)
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

        # 무한 주문 반복
        m = -1
        while m != 0:
            
            #음료 번호 선택
            n = 0
            while n < 1 or len(menu) < n:
                n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력
    
                # 메뉴판에 있는 음료 번호일때
                if 1 <= n & n <= len(menu):
                    self.price_sum = self.price[n-1]  # 선택 음료의 가격
            
                # 메뉴판에 없는 번호일 때
                else:  
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
    
            # 음료 온도 물어보기
            t = 0  # 기본값
            while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                t = int(input("HOT 음료는 '1'을, ICE 음료는 '2'를 입력하세요 : "))       
                if t == 1:
                    self.temp = "HOT" #self 없으면 메서드안에서만 이용, self 붙이면 클래스 전체 이용하는 인스턴스 변수로 인
                elif t == 2:
                    self.temp = "ICE"
                else:    
                    print("1과 2 중 하나를 입력하세요.")
            
            print('[주문 음료', self.temp, n, self.menu[n-1], ':', self.price_sum, '원]')

            # (위치가 이동되었습니다) 주문 리스트에 추가
            self.order_price.append(self.price[n-1])  # 가격 리스트에 추가
            self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가

            # 추가 주문 물어보기
            m = -1
            while m != 0 and m != 1:   # 0이나 1를 입력할 때까지 물어보기
                m = int(input("\n추가 주문은 '1'을, 주문 완료 후 결제는 '0'을 입력하세요 : "))
                if m == 0:
                    print("주문이 완료되었습니다.\n")
                elif m != 0 and m != 1:
                    print("0과 1 중 하나를 선택하세요.\n")

        
        #주문 완료 시 리스트 출력
        self.price_sum = sum(self.order_price)
        print(f"주문 메뉴: {self.order_menu}")
        print(f"주문 금액: {self.order_price}")

    #지불 메소드
    def pay(self):
        print(f"*** 지불하실 총 금액은 {self.price_sum}원 입니다. ***")

        #지불 방법
        payment = 0
        while payment != 1 and payment != 2:
            payment = int(input("\n현금 결제는 '1', 카드 결제는 '2'를 입력해주세요."))
            if payment == 1:
                print("직원을 호출하겠습니다.\n")
            elif payment == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.\n")
            else:
                print("다시 결제를 시도해주세요. 1 또는 2만 입력이 가능합니다.")

    #주문표 메소드
    def table(self):
        #외곽 출력
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        #주문 내역 나열; 상품명 : 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        #합계 금액       
        print('합계 금액 :', self.price_sum)

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()
```
