```python
import random

class Account:
    # Q2. 클래스 변수: 생성된 계좌의 개수 저장
    account_count = 0

    # Q1. 생성자 구현(예금주, 초기 잔액 입력 받기)
    def __init__(self, name, balance):
        self.bank = "SC은행"
        self.name = name
        self.balance = balance

        # 계좌번호 랜덤 생성 (11자리)
        num1 = str(random.randint(10**10, 10**11 - 1))
        self.account_number = f"{num1[0:3]}-{num1[3:5]}-{num1[5:]}"

        # 계좌가 생성될 때마다 클래스 변수 1씩 증가
        Account.account_count += 1

        # Q7을 위한 입금 횟수 카운트 변수 추가
        self.deposit_count = 0

        # 10을 위한 입금 입출금 내역 로그 변수 추가
        self.deposit_log = []
        self.withdraw_log = []

    # Q3. 클래스 변수 출력 메서드
    def get_account_num(self):
        return f"현재 생성된 계좌 개수: {Account.account_count}개"

    # Q4. 입금 메서드
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            print(f"{amount}원이 입금되었습니다.")

            # 입금 내역 로그에 기록
            temp_dep = str(format(amount,','))
            self.deposit_log.append(f"{temp_dep}원")

            # Q7. 입금 횟수가 5의 배수일 때마다 1% 이자 지급
            self.deposit_count += 1
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"** 5회 입금하시어 '이자 {interest}원' 지급 완료! **")

        else:
            print("입금은 1원 이상만 가능합니다.")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount}원이 출금되었습니다.")

            # 출금 내역 로그에 기록
            temp_wit = str(format(amount,','))
            self.withdraw_log.append(f"{temp_wit}원")

        else:
            print("잔액이 부족합니다.")

    # Q6. 정보 출력 메서드
    def display_info(self):
        # 잔고 세자리마다 쉼표 추가: format(숫자, ',') or f"{balance:,}"
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.name}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {format(self.balance, ',')}원")
        print("-" * 20)

    # Q.10 입금 내역 출력 메서드
    def deposit_history(self):
        print(f"\n** 입금 내역은 다음과 같습니다. **")
        print(f"{self.deposit_log}")

    # Q.10 출금 내역 출력 메서드
    def withdraw_history(self):
        print(f"\n** 출금 내역은 다음과 같습니다. **")
        print(f"{self.deposit_log}")

```

```python
# Q8. 3개 객체 생성 및 리스트 저장
person1 = Account("최유현", 100000000000000000)
person2 = Account("콩이", 20000000)
person3 = Account("치치", 500000)
person4 = Account("뽀또", 9000)

account_list = [person1, person2, person3, person4]
```

```python
# 계좌 확인
for a in account_list:
    a.display_info()

print(person1.get_account_num())
```

```python
person4.deposit(600000)
person4.deposit(100000)
person4.deposit(100000)
person4.deposit(100000)
person4.deposit(100000)

person4.deposit_history()
```

```python
# Q9. 100만원 이상의 잔고가 있는 고객 정보만 출력
def VIP_person():
  print("** 우리 은행 VIP고객 정보는 아래와 같습니다. **\n")

  for a in account_list:
    if a.balance >= 1000000:
      a.display_info()

VIP_person()
```
